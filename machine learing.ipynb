```python
import pandas as pd
import numpy as np
import os
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier, VotingClassifier, StackingClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.impute import SimpleImputer
from sklearn.feature_selection import VarianceThreshold, SelectKBest, f_classif
from sklearn.preprocessing import StandardScaler
from imblearn.over_sampling import SMOTE
import matplotlib.pyplot as plt
import seaborn as sns

# ===== 参数配置 =====
currency_pairs = [
    'CNY_USD', 'CNY_GBP', 'USD_GBP',
    'USD_CNY', 'GBP_CNY', 'GBP_USD'
]
bond_files = {
    'CN': 'china_bond_simple.csv',
    'US': 'us_bond_simple.csv',
    'UK': 'uk_bond_simple.csv'
}
news_feat_csv = 'news_features.csv'
random_seed = 42
select_k = 10  # 特征选择数量
lookahead = 3  # 标签优化：未来N天
threshold = 0.002  # 标签涨幅阈值（如0.2%）

# ===== 读取国债数据 =====
bond_data = {}
for country, file in bond_files.items():
    df = pd.read_csv(file)
    df['timestamp'] = pd.to_datetime(df['timestamp'])
    df = df.rename(columns={'close': f'bond_{country}'})[['timestamp', f'bond_{country}']]
    bond_data[country] = df

# ===== 读取新闻特征 =====
news_feat = pd.read_csv(news_feat_csv)
news_feat['timestamp'] = pd.to_datetime(news_feat['day'], format='%Y%m%d')
news_feat_pivot = news_feat.pivot(index='timestamp', columns='country', values=['news_count', 'avg_tone'])
news_feat_pivot.columns = ['_'.join(col).strip() for col in news_feat_pivot.columns.values]
news_feat_pivot = news_feat_pivot.reset_index()

# ===== 定义基础模型列表 =====
base_models = [
    ('lr', LogisticRegression(max_iter=1000, random_state=random_seed)),
    ('rf', RandomForestClassifier(n_estimators=100, random_state=random_seed, n_jobs=-1)),
    ('gb', GradientBoostingClassifier(n_estimators=100, random_state=random_seed))
]
other_models = {
    'KNeighbors': KNeighborsClassifier(n_neighbors=15, n_jobs=-1),
    'SVM': SVC(probability=True, random_state=random_seed)
}

# ====== 调参空间（可根据需要自行扩展）======
rf_param_grid = {
    'n_estimators': [100, 200],
    'max_depth': [3, 5, None],
    'min_samples_split': [2, 5]
}
gb_param_grid = {
    'n_estimators': [100, 200],
    'learning_rate': [0.05, 0.1],
    'max_depth': [3, 5]
}

results = []

for pair in currency_pairs:
    fx_file = f"{pair}.csv"
    if not os.path.exists(fx_file):
        print(f"File not found: {fx_file}, skipped.")
        continue

    print(f"\n===== {pair} =====")
    fx = pd.read_csv(fx_file)
    fx['timestamp'] = pd.to_datetime(fx['timestamp'])
    fx = fx[['timestamp', 'close']].sort_values('timestamp')

    # 合并国债数据
    for country in bond_data:
        fx = fx.merge(bond_data[country], on='timestamp', how='left')
    # 合并新闻特征
    fx = fx.merge(news_feat_pivot, on='timestamp', how='left')

    # ===== 标签优化：未来N天涨幅阈值 =====
    fx['future_close'] = fx['close'].shift(-lookahead)
    fx['target'] = ((fx['future_close'] - fx['close']) / fx['close'] > threshold).astype(int)
    fx = fx.dropna(subset=['target'])

    # 特征与标签
    exclude_cols = ['timestamp', 'close', 'future_close', 'target']
    feature_cols = [col for col in fx.columns if col not in exclude_cols]
    X = fx[feature_cols]
    y = fx['target']

    # 1. 填充缺失值
    imputer = SimpleImputer(strategy='mean')
    X_imputed = imputer.fit_transform(X)

    # 2. 方差过滤
    vt = VarianceThreshold(threshold=1e-5)
    X_vt = vt.fit_transform(X_imputed)

    # 3. 特征选择（选择K个最相关）
    k_select = min(select_k, X_vt.shape[1])
    skb = SelectKBest(f_classif, k=k_select)
    X_selected = skb.fit_transform(X_vt, y)

    # 4. 标准化
    scaler = StandardScaler()
    X_scaled = scaler.fit_transform(X_selected)

    # 5. 切分训练/测试集
    split_idx = int(len(X_scaled) * 0.8)
    X_train, X_test = X_scaled[:split_idx], X_scaled[split_idx:]
    y_train, y_test = y.iloc[:split_idx], y.iloc[split_idx:]

    print("Train label counts:", np.bincount(y_train.astype(int)))
    print("Test label counts:", np.bincount(y_test.astype(int)))

    # ===== 样本平衡（SMOTE过采样） =====
    smote = SMOTE(random_state=random_seed)
    X_train_bal, y_train_bal = smote.fit_resample(X_train, y_train)
    print("After SMOTE:", np.bincount(y_train_bal.astype(int)))

    # ====== 模型调参（以RF/GB为例）======
    rf_grid = GridSearchCV(RandomForestClassifier(random_state=random_seed, n_jobs=-1), rf_param_grid, cv=3, scoring='accuracy', n_jobs=-1)
    rf_grid.fit(X_train_bal, y_train_bal)
    print("Best RF params:", rf_grid.best_params_)
    best_rf = rf_grid.best_estimator_

    gb_grid = GridSearchCV(GradientBoostingClassifier(random_state=random_seed), gb_param_grid, cv=3, scoring='accuracy', n_jobs=-1)
    gb_grid.fit(X_train_bal, y_train_bal)
    print("Best GB params:", gb_grid.best_params_)
    best_gb = gb_grid.best_estimator_

    # Voting和Stacking集成用调优后模型
    voting = VotingClassifier(estimators=[
        ('lr', LogisticRegression(max_iter=1000, random_state=random_seed)),
        ('rf', best_rf),
        ('gb', best_gb)
    ], voting='soft')
    stacking = StackingClassifier(estimators=[
        ('lr', LogisticRegression(max_iter=1000, random_state=random_seed)),
        ('rf', best_rf),
        ('gb', best_gb)
    ], final_estimator=LogisticRegression())

    all_models = {
        'Voting': voting,
        'Stacking': stacking,
        'LogisticRegression': LogisticRegression(max_iter=1000, random_state=random_seed),
        'RandomForest': best_rf,
        'GradientBoosting': best_gb,
        'KNeighbors': KNeighborsClassifier(n_neighbors=15, n_jobs=-1),
        'SVM': SVC(probability=True, random_state=random_seed)
    }

    pair_results = {}

    # 修复特征名索引
    feature_names_step1 = np.array([feature_cols[i] for i in vt.get_support(indices=True)])
    feature_names = feature_names_step1[skb.get_support(indices=True)]
    feature_importances_dict = {}

    for name, model in all_models.items():
        if len(np.unique(y_train_bal)) < 2 or len(np.unique(y_test)) < 2:
            print(f"[{name}] Skipped: Not enough classes in train or test set.")
            pair_results[name] = np.nan
            continue
        print(f"[{name}] Training...")
        model.fit(X_train_bal, y_train_bal)
        y_pred = model.predict(X_test)
        acc = accuracy_score(y_test, y_pred)
        pair_results[name] = acc
        print(f"Test Accuracy: {acc:.4f}")

        # 特征重要性分析（仅树模型）
        if hasattr(model, 'feature_importances_'):
            importances = model.feature_importances_
            feature_importances_dict[name] = importances

    # 可视化对比
    plt.figure(figsize=(12, 4))
    plt.bar(pair_results.keys(), pair_results.values(), color='skyblue')
    plt.title(f'Model Accuracy Comparison for {pair}')
    plt.ylabel('Accuracy')
    plt.ylim(0, 1)
    plt.xticks(rotation=20)
    plt.grid(axis='y', linestyle='--', alpha=0.6)
    plt.show()

    # 详细报告
    if len(np.unique(y_test)) > 1 and 'RandomForest' in all_models:
        print("\nClassification Report (RandomForest):")
        y_pred_rf = all_models['RandomForest'].predict(X_test)
        print(classification_report(y_test, y_pred_rf))
        print("Confusion Matrix:")
        print(confusion_matrix(y_test, y_pred_rf))

    # 特征重要性可视化（随机森林/GBDT）
    for mdl_name, importances in feature_importances_dict.items():
        indices = np.argsort(importances)[::-1]
        plt.figure(figsize=(8, 5))
        plt.title(f"{mdl_name} Feature Importances ({pair})")
        plt.bar(range(len(indices)), importances[indices], align="center")
        plt.xticks(range(len(indices)), feature_names[indices], rotation=45, ha='right')
        plt.tight_layout()
        plt.show()

    pair_results['pair'] = pair
    results.append(pair_results)

# ===== 汇总所有货币对结果 =====
results_df = pd.DataFrame(results).set_index('pair')
print("\n==== Model Accuracy Comparison Across All Currency Pairs ====")
print(results_df)

plt.figure(figsize=(12, 5))
sns.heatmap(results_df, annot=True, fmt=".3f", cmap='Blues')
plt.title("Model Accuracy Comparison Across All Currency Pairs")
plt.show()
# 训练好模型后加上这一行
print("本模型训练用的特征名：")
print(list(feature_names))
```
